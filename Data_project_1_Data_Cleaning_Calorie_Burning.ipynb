Removing null data

2.   Remove Null Values


> For analyzing data, empty cell should be removed.


```python
import pandas as pd
url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/Data%20Cleaning%20Data%20Sets/dirtydata.csv"
df = pd.read_csv(url)
## Return a new Data Frame with no empty cells
new_df=df.dropna()
new_df
# Result should show, row 18,22,28 removed
```

```python
# Fill null values data with 0
df.fillna(0)
```

```python
## Replace NULL values in the "Calories" columns with the number

df["Calories"].fillna(150, inplace = True)
print(df.to_string())

# Result should show row 18, Calories value= 150
```

```python
#Calculating the MEAN, and replace any empty values with it

x = df["Calories"].mean()
df["Calories"].fillna(x, inplace = True)
print(df.to_string())

# Result should show row 18, Calories value= Mean value
```

```python
## Input missing values: Adding Date values ##
# convert all cells in the 'Date' column into dates
df['Date'] = pd.to_datetime(df['Date'])
print(df.to_string())

```

Removing null data


> The result from the converting in the Date above gave a NaT value, which can be handled as a NULL value, and the row can be removed by using the dropna() method.


```python
df['Date'] = pd.to_datetime(df['Date'])

df.dropna(subset=['Date'], inplace = True)

print(df.to_string())
## row 22 will be removed
```

Removing wrong Data

> Duration coloumn data is between 30 to 60. Row 7 shows 450 which is abrupt data.
To fix wrong values is to replace them with something else, row 7 Duration value can be replaced by "45" considering "450" as typo mistake


```python
# Removing wrong data
# row 7 Duration data 450 replaced with data 45
df.loc[7,'Duration'] = 45
print(df.to_string())

# Result should show row 7 Duration data 450 replaced with 45
```

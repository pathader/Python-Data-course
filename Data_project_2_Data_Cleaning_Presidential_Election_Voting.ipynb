**Removing null data**

2.   Remove Null Values


> For analyzing data, empty column should be removed. Here in the code removing "national_voteshare_3rd_lo" using drop() method



```python
# Removing column "national_voteshare_3rd_lo" using drop()
import pandas as pd
from datetime import datetime, timezone
url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/Data%20Cleaning%20Data%20Sets/presDirty.csv"
data_frame = pd.read_csv(url)

data_frame.drop('national_voteshare_3rd_lo', axis=1, inplace=True)
data_frame
```

**Removing column "cycle" and "simulations" using drop() method**

```python
# Removing column "cycle" and "simulations using drop() method"
data_frame.drop('cycle', axis=1, inplace=True)
data_frame.drop('simulations', axis=1, inplace=True)
data_frame
```

**Fill null values data with 0, use df.fillna(0)**

```python
# Fill null values data with 0
data_frame.fillna(0)
```

**In the row 60, "modeldate" column doesn't have value. Considering row 59 and 61 "modeldate" column values,in row 60 date is added as '09/04/2020'**

```python
## Input missing values: Adding Date values ##
data_frame.loc[60,'modeldate'] = '09/04/2020'
#print(data_frame.to_string())
#data_frame
print(data_frame.loc[60,'modeldate'])
```

**Converting the date format to dd/mm/yyyy**

```python
# converting the date format to dd/mm/yyyy

# Convert 'modeldate' column to datetime format
data_frame['modeldate'] = pd.to_datetime(data_frame['modeldate'])

# Change date format to dd/mm/yyyy
data_frame["modeldate"] = data_frame["modeldate"].dt.strftime('%d/%m/%Y')

print(data_frame["modeldate"])
```

**Inserting missing value "Trump" in the column "candidate_inc"**

```python
# inserting missing value "Trump" in the column "candidate_inc"

data_frame.loc[8,'candidate_inc'] = 'Trump'
data_frame.loc[19,'candidate_inc'] = 'Trump'

#print(data_frame.to_string())
print(data_frame["candidate_inc"])
```

**Replacing empty values of the columns national_turnout, national_turnout_hi, national_turnout_lo**

```python
#Calculating the MEAN, and replace any empty values with it

# Replacing Null value of national_turnout by mean
x = data_frame["national_turnout"].mean()

data_frame["national_turnout"].fillna(x, inplace=True)

#Replacing Null value of national_turnout_hi by mean
y = data_frame["national_turnout_hi"].mean()
data_frame["national_turnout_hi"].fillna(y, inplace = True)


# Replacing Null value of national_turnout_lo by mean
z = data_frame["national_turnout_lo"].mean()
data_frame["national_turnout_lo"].fillna(z, inplace = True)


#print(data_frame.to_string())
print(data_frame[["national_turnout", "national_turnout_hi", "national_turnout_lo"]])

```
